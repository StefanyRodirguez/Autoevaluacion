```python
import requests#importa la biblioteca de solicitudes 
from bs4 import BeautifulSoup #importa la libreria beautifulsoup 


page = requests.get("https://www.yelp.com/biz/milk-and-cream-cereal-bar-new-york?osq=Ice+Cream")
page
page.content
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
comentarios = soup.findAll("span", {"class":"lemon--span__373c0__3997G raw__373c0__3rcx7"})
comentarios
```

```python
comen = list() 

for i in comentarios:
    comen.append(i.text)
print(comen)
```
